**1.1. csv'i okuma+ ilk 5 satıra bakma**

```python
import pandas as pd
df = pd.read_csv('veri_seti_200k.csv', sep=';', on_bad_lines='skip') #veri setini çektik ve ; ile ayırdık her bir satırı
                                                                     #bad lines'ları yani bozuk satırları atladık
df.head() # ilk 5 satırı göster sadece
```

1.2. eksik veri kontrolü

```python
print("--- Eksik Veri Sayıları ---")
print(df.isnull().sum()) # df.isnull() ile satırlardaki NaN değerlerini saydık
```

1.3. temel istatistikler

```python
print("--- Temel İstatistikler ---")
print(df.describe()) # Veri setindeki sayısal sütunları analiz ediyor

# Veri setinin genel yapısını daha teknik detaylarını analiz eder
print("\n--- Veri Seti Özeti ---")
print(df.info())
```

```python
## 2.1. küçük harfe çevirme
```

```python
df['kucuk harfe cevirme'] = df['Metin'].str.lower()

# Eski hali ve küçük harfli yeni halini yan yana görelim
df[['Metin', 'kucuk harfe cevirme']].head()
```

2.2. noktalama işaretleri

```python
# [^\w\s] şu demek: Harfler (\w) ve boşluklar (\s) HARİCİNDEKİ (^) her şeyi bul ve sil ('')
df['Noktalama Yok'] = df['kucuk harfe cevirme'].str.replace(r'[^\w\s]', '')

df[['Metin', 'kucuk harfe cevirme', 'Noktalama Yok']].head()
```

2.3. sayıları kaldırma

```python
# \d  ile tüm sayıları (rakamları) bul ve hiçbir şeyle ('') değiştir, yani sil. \d+ tüm sayılar demek
df['Sayisiz'] = df['Noktalama Yok'].str.replace(r'\d', '')

df[['Metin', 'kucuk harfe cevirme', 'Noktalama Yok', 'Sayisiz']].head()
```

2.4. fazla boşlukları sil

```python
# \s+ şifresi: "Yan yana duran birden fazla boşluğu (\s) yakala" demek.
# .str.strip() ise cümlenin taa en başındaki ve en sonundaki (varsa) boşlukları kırpar.

df['Tamamen Temiz'] = df['Sayisiz'].str.replace(r'\s+', ' ').str.strip()

df[['Metin', 'Tamamen Temiz']].head(10)
```

3.1 kelime ayırma

```python
df['Token'] = df['Tamamen Temiz'].str.split()

df[['Tamamen Temiz', 'Token']].head()
```

3.2. stop word temizleme

```python
stop_words = ['fakat','lakin','ancak','acaba', 'ama', 'aslında', 'az', 'bazı', 'belki', 'biri', 'birkaç', 'birşey', 'biz', 'bu', 'çok', 'çünkü', 'da', 'daha', 'de', 'defa', 'diye', 'eğer', 'en', 'gibi', 'hem', 'hep', 'hepsi', 'her', 'hiç', 'için', 'ile', 'ise', 'kez', 'ki', 'kim', 'mı', 'mu', 'mü', 'nasıl', 'ne', 'neden', 'nerde', 'nerede', 'nereye', 'niçin', 'niye', 'o', 'sanki', 'şey', 'siz', 'şu', 'tüm', 've', 'veya', 'ya', 'yani']

df['Filtreli_Token'] = df['Token'].apply(lambda kelimeler: [kelime for kelime in kelimeler if kelime not in stop_words])

# Elekten önceki hali ve sonraki halini yan yana görelim
df[['Token', 'Filtreli_Token']].head(10)
```

3.3. en çok tekrar eden kelimeler

```python
# tüm kelimeleri tek tek sütun yapıyoruz
tum_kelimeler = df['Filtreli_Token'].explode()

# her kelimeden kaç tane var
kelime_sayilari = tum_kelimeler.value_counts()

#en çok tekrar eden lk 20 kelimeyi yazdırıyo
print("İLK 20 KELİME")
print(kelime_sayilari.head(20))
```

4.1.bar chart (20 kelime)


```python
import matplotlib.pyplot as plt
import seaborn as sns

# ilk 20 kelimeyi seç
ilk_20 = kelime_sayilari.head(20)

plt.figure(figsize=(10, 10))

# x eksenine values, y eksenine index koyuyoruz
sns.barplot(x=ilk_20.values, y=ilk_20.index) #seaborn kütüphanesi

# Grafiğimize başlık ve etiketler ekliyoruz
plt.title('En Sık Kullanılan 20 Kelime', fontsize=16) #matplotlib
plt.xlabel('Kullanım Sayısı', fontsize=12)
plt.ylabel('Kelimeler', fontsize=12)

# Ekranda göster!
plt.show()
```

4.2.Histogram: Yorum uzunlukları

```python
df['Yorum_Uzunlugu'] = df['Metin'].apply(len)

plt.figure(figsize=(15, 10))

sns.histplot(df['Yorum_Uzunlugu'], bins=50, color='purple') #bins: Verileri kaç tane sütuna/kutucuğa böleceği

plt.title('Yorum Uzunluklarının Dağılımı', fontsize=16)
plt.xlabel('Bir Yorumdaki Kelime Sayısı', fontsize=12)
plt.ylabel('Yorum Sayısı (Frekans)', fontsize=12)

plt.show()
```

4.3.Pie chart: Kelime dağılımları

```python
ilk_10 = kelime_sayilari.head(10)

plt.figure(figsize=(10, 10))
plt.pie(ilk_10.values,
        labels=ilk_10.index,
        autopct='%1.1f%%',
        startangle=140)
plt.title('En Sık Kullanılan İlk 10 Kelimenin Dağılımı', fontsize=16, fontweight='bold')

plt.show()
```

4.4. Kelime bulutu (wordcloud kütüphanesi - tek ek kütüphane)

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

tum_metin = " ".join(df['Tamamen Temiz'].astype(str))
wordcloud = WordCloud(width=800,
                      height=400,
                      background_color='white',
                      max_words=100).generate(tum_metin)

plt.figure(figsize=(15, 10))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python

tum_kelimeler_listesi = [kelime for alt_liste in df['Filtreli_Token'] for kelime in alt_liste]

toplam_kelime_sayisi = len(tum_kelimeler_listesi)

benzersiz_kelime_sayisi = len(set(tum_kelimeler_listesi))

cesitlilik_orani = benzersiz_kelime_sayisi / toplam_kelime_sayisi

print(f"Toplam Kelime Sayısı: {toplam_kelime_sayisi}")
print(f"Benzersiz Kelime Sayısı: {benzersiz_kelime_sayisi}")
print(f"Kelime Çeşitliliği Oranı: {cesitlilik_orani:.4f}")
```
